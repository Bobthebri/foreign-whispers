### 5.1 — Design your voice assignment strategy

Some options to consider:
- **Round-robin:** Assign voices in order (SPEAKER_00 → voice_1.wav, SPEAKER_01 → voice_2.wav)
- **Filename-based:** Match speaker index to file index
- **Default + override:** Use `default.wav` for all speakers unless specific speaker WAVs are provided

Document your chosen approach in the cell below:

### Voice Assignment Strategy (Task 5)

**Strategy chosen:** Round-robin by index with per-language default fallback

**How speakers map to voice files:**
- Resolution order (implemented in `foreign_whispers/voice_resolution.py`):
  1. `speakers/{lang}/{speaker_id}.wav` — if a file named `SPEAKER_00.wav` etc. exists
  2. `speakers/{lang}/default.wav` — language-specific default for all other speakers
  3. `speakers/default.wav` — global fallback when no language directory exists
- The TTS router reads `speaker` fields from the translated JSON segments, collects unique speaker IDs, then calls `resolve_speaker_wav` for each, building a `voice_map: dict[str, str]`.
- `voice_map` is passed to `TTSService.text_file_to_speech` → `tts_text_file_to_speech`, which switches reference audio per segment during synthesis.

**Edge cases considered:**
- No speaker labels in transcript → `voice_map` is `None`, single default voice used
- Only one speaker detected → all segments use `default.wav`
- Unknown speaker ID with no matching WAV → falls back to `default.wav`
- Missing `speakers/` directory → `resolve_speaker_wav` returns `"default.wav"` gracefully
